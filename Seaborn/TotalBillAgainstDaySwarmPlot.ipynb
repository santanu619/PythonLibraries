```python
'''
@Author: Santanu Mohapatra
@Date:21/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 15:50 PM
@Title: Python program to create a swarm plot for total bill against day using tips dataset.
'''

import matplotlib.pyplot as plt 
import seaborn as sns 

tips_dataset = sns.load_dataset("tips")

sns.swarmplot(x = 'total_bill', y = 'day', data = tips_dataset)
plt.show()
```
