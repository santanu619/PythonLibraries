```python
'''
@Author: Santanu Mohapatra
@Date:21/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 13:10 PM
@Title: Python program to create a violin plot using tips csv dataset for gender vs days.
'''

import matplotlib.pyplot as plt 
import seaborn as sns 

tips_dataset = sns.load_dataset("tips")

sns.violinplot(x = 'day', y = 'tip', hue = 'sex', data = tips_dataset)
plt.show()
```
