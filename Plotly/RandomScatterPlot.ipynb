```python
'''
@Author: Santanu Mohapatra
@Date:21/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 18:10 PM
@Title: Python program to draw a scatter plot for random 1000 x and y coordinates.
'''

import plotly.express as px
import numpy as np 

number = 1000
x = np.random.randn(number)
y = np.random.randn(number)

fig = px.scatter(x, y)
fig.show()
```
