```python
'''
@Author: Santanu Mohapatra
@Date: 18/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 15:30 PM
@Title: Python program to show and change the axis values.
'''
import matplotlib.pyplot as plt

X = [2,3,4,8]
Y = [4,6,8,16]
print("Values of X:")
print(X) 
print("Values of Y:")
print(Y)
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Straight Line')
print(plt.axis())
plt.axis([0,20,0,20])
plt.show()
```
