```python
'''
@Author: Santanu Mohapatra
@Date: 18/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 14:50 PM
@Title: Python program to plot graphs using line marker.
'''
import matplotlib.pyplot as plt

x1 = [2,5,6]
y1 = [3,2,1]
plt.plot(x1,y1,label = "line 1",linestyle = 'dashed', marker = 'o', markerfacecolor = 'green', markersize = 10)

x2 = [2,3,4]
y2 = [3,4,2]
plt.plot(x2,y2,label = "line 2",linestyle = 'dotted',marker = 'o',markerfacecolor = 'yellow', markersize = 10)

plt.xlim(1,8)
plt.ylim(1,8)

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title("PLOTS WITH LINE MARKERS")

plt.legend()
plt.show()
```
