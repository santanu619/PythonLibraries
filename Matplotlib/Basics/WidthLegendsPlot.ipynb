```python
'''
@Author: Santanu Mohapatra
@Date: 18/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 13:20 PM
@Title: Python program to plot two different lines including legend and different widths.
'''
import matplotlib.pyplot as plt

x1 = [12,18,24]
y1 = [10,20,30]
plt.plot(x1,y1,label = "line 1",linewidth = 4)

x2 = [10,20,30]
y2 = [15,30,12]
plt.plot(x2,y2,label = "line 2",linewidth = 2)

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title("PLOTS WITH LEGENDS AND WIDTHS")

plt.legend()
plt.show()
```
