```python
'''
@Author: Santanu Mohapatra
@Date: 18/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 12:20 PM
@Title: Python program to plot a straight line.
'''
import matplotlib.pyplot as plt
import numpy as np 
X = np.arange(0,10)
Y = X*2
print("Values of X:")
print(X) 
print("Values of Y:")
print(Y)
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Straight Line')
plt.show()
```
