```python
'''
@Author: Santanu Mohapatra
@Date: 19/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 18:20 PM
@Title: Python program to create a stacked bar plot with error bars.
'''

import numpy as np
import matplotlib.pyplot as plt

number = 5
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStandardDeviation = (4, 3, 4, 1, 5)
womenStandardDeviation = (3, 5, 2, 3, 3)

index = np.arange(number)    
width = 0.35      

plot1 = plt.bar(index, menMeans, width, yerr=menStandardDeviation, color='yellow')
plot2 = plt.bar(index, womenMeans, width,
bottom=menMeans, yerr=womenStandardDeviation, color='blue')

plt.ylabel('Scores')
plt.xlabel('Groups')
plt.title('Stacked Bar Plot')
plt.xticks(index, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((plot1[0], plot2[0]), ('Men', 'Women'))

plt.show()
```
