```python
'''
@Author: Santanu Mohapatra
@Date: 19/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 17:40 PM
@Title: Python program to display a bar chart from the data frame.
'''

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

myarray = np.array([[4,8,5,7,6],[2,3,4,2,6],[4,7,4,7,8],[2,6,4,8,6],[2,4,3,3,2]])
data_frame = pd.DataFrame(myarray, columns=['a','b','c','d','e'], index=[2,4,6,8,10])

data_frame.plot(kind='bar')

plt.show()
```
