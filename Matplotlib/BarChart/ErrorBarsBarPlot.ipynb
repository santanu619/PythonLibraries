```python
'''
@Author: Santanu Mohapatra
@Date: 19/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 17:50 PM
@Title: Python program to create bar plots with error bar on the same plot.
'''

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

N = 5
men_means = (54.74, 42.35, 67.37, 58.24, 30.25)
men_std= (4, 3, 4, 1, 5)

ind = np.arange(N)  
width = 0.35       

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='r', yerr=men_std)


plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')

red_patch = mpatches.Patch(color='red', label='Men')
plt.legend(handles=[red_patch])


def autolabel(rects):
    """
    Description: Attach a text label above each bar displaying its height.
    """
    for rect in rects:
       height = rect.get_height()
       ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
       ha='center', va='bottom')
autolabel(rects1)
plt.show()
```
