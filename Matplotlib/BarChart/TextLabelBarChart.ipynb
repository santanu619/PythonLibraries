```python
'''
@Author: Santanu Mohapatra
@Date: 19/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 13:10 PM
@Title: Python program to display a bar chart of the popularity of the programming language displaying text label.
'''
import matplotlib.pyplot as plt

programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
programming_languages_pos = [index for index, _ in enumerate(programming_languages)]

bar_plot = plt.bar(programming_languages_pos, popularity, color='green', edgecolor='red')

def autolabel(rects):
    for index,rect in enumerate(bar_plot):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                popularity[index],
                ha='center', va='bottom', rotation=0)

autolabel(bar_plot)

plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language")

plt.show()
```
