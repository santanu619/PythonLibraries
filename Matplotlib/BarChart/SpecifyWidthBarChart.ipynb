```python
'''
@Author: Santanu Mohapatra
@Date: 19/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 14:10 PM
@Title: Python program to display a bar chart of the popularity of the programming language giving specified widths.
'''
import matplotlib.pyplot as plt

programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
programming_languages_pos = [index for index, _ in enumerate(programming_languages)]
y_pos = [0,1,4,5,8,9]

plt.bar(y_pos, popularity, edgecolor='red', width=[0.1,0.4,0.7,0.2,0.5,0.2])

plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language")

plt.show()
```
