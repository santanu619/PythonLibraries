```python
'''
@Author: Santanu Mohapatra
@Date: 20/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 12:10 PM
@Title: Python program to create a pie chart of gold medal achievements reading from csv files.
'''

import matplotlib.pyplot as plt
import pandas as pd

dataframe  =  pd.read_csv('medal.csv')
country_data = dataframe["country"]
medal_data = dataframe["gold_medal"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0, 0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Gold Medal Achievements", bbox = {'facecolor':'0.8','pad':5})
plt.show()
```
