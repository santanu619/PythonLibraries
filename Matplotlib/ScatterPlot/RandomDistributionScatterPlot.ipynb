```python
'''
@Author: Santanu Mohapatra
@Date: 20/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 12:30 PM
@Title: Python program to draw a scatterplot taking random distributions.
'''

import math
import random
import matplotlib.pyplot as plt
# create random data
number = 100
x = [random.triangular() for index in range(number)]
y = [random.gauss(0.5, 0.25) for index in range(number)]
colors = [random.randint(1, 4) for index in range(number)]
areas = [math.pi * random.randint(5, 15)**2 for index in range(number)]
# draw the plot
plt.figure()
plt.scatter(x, y, s=areas, c=colors, alpha=0.85)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
