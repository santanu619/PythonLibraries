```python
'''
@Author: Santanu Mohapatra
@Date: 20/06/2021
@Last Modified by: Santanu Mohapatra
@Last Modified Time: 13:10 PM
@Title: Python program to draw a scatterplot taking random distributions generating balls of different sizes.
'''

import matplotlib.pyplot as plt 
import math
import random
import numpy as np 

x = np.random.randn(50) 
y = np.random.randn(50)
areas = [math.pi * random.randint(5, 15)**2 for i in range(0,50)]
plt.scatter(x, y, s=areas, facecolors='g')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
